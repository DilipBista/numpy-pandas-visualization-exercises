```python
import matplotlib.pyplot as plt
import math 
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

plt.suptitle("Big O Notation")

n = np.linspace(1, 20, 800)
plt.ylim(0, 100)



labels = ['0(1)', '0(log n)', '0(n)', '0(n log n)', '0(n^2)', '0(2^n)', '0(n^n)']
big_o = [np.ones(n.shape), np.log(n), n, n * np.log(n), n**2, 2**n, n**n]

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])
    
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()


```
