4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python

def get_db_url(db_name):
    from env import user, host, password
    return f'mysql+pymysql://{user}:{password}@{host}/{db_name}'
```

```python
sql1 = """ SELECT * from orders"""
```

```python
url1 = get_db_url("chipotle")

df1 = pd.read_sql(sql1, url1)
df1
```

```python
df1['price']=df1.item_price.str.replace('$','').astype(float)
df1
```

```python
popular4= df1.groupby('item_name').quantity.sum().nlargest(4)
popular4
```

```python
df1
```

```python
popular4.index
```

```python
df1.item_name.isin(popular4.index)
```

```python
sum_price= df1[df1.item_name.isin(popular4.index)].groupby('item_name').price.sum()
sum_price
```

```python
sum_price = sum_price.sort_values(ascending = False).reset_index()
sum_price
```

```python
plt.figure(figsize= (6,4))
sns.barplot(y = 'item_name', x = 'price', data = sum_price)
plt.title('4 Most popular')
```
