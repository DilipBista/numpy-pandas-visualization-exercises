```python
'''  Use matplotlib to plot the following equation:

y = x^2 - x+2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.''' 
```

```python
import matplotlib.pyplot as plt 
import numpy as np
# generate data

x = range (-50,50)
y =[x**2 -x+2 for x in x]
# plot x & y
plt.plot(x,y)

#add label 
plt.xlabel("x axis")
plt.ylabel("y axis")
# add annotation 
plt.annotate('Origin', xy = (0,0), xytext = (0, 1000), arrowprops ={})

plt.show()
```

```python
''' Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y = x^1/2
y= x**3
y= 2^x
y = 1/(x+1)
You can use functions from the math module to help implement some of the equations above.'''
```

```python
x1 = range(1, 60)
y1 = [n**(1/2) for n in x1]
plt.plot(x1,y1)

plt.title(" $y = \sqrt{x}$")
plt.xlabel(" x axis")
plt.ylabel("y axis")
```

```python
x2 = range(1,15)
y2 = [n**3 for n in x2]
plt.plot(x2,y2)
plt.title(" $y = x^3$")
plt.xlabel(" x axis")
plt.ylabel("y axis")
```

```python
x3 = range(1, 120)
y3 = [2**n for n in x3 ]
plt.plot(x3,y3)
plt.title(" $y = 2^x$")
plt.xlabel(" x axis")
plt.ylabel("y axis")
```

```python
x4 = range(1, 6)
y4 = [1/ (n+1) for n in x4]
plt.plot(x4,y4)
plt.title (" y = 1/(x+1)")
plt.xlabel(" x axis")
plt.ylabel("y axis")
```
